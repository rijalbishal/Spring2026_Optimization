```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

t= np.linspace(0, 2*np.pi ,400)
x1e= 3*np.cos(t)
x2e=2*np.sin(t)

x1=np.linspace(-4,4,2000)
x2= (3- x1**2)/3

feasible_mask = (x1>= -1) & (4*x1**2 + 9*x2**2 <= 36) 
x1_feas = x1[feasible_mask]
x2_feas = x2[feasible_mask]

plt.figure()                         
plt.plot(x1e,x2e,label='4x1^2+9x2^2<=36')
plt.plot(x1,x2,label='x1^2+3x2 =3')
plt.plot(x1_feas,x2_feas,label='Feasible region')
plt.axvline(-1,label='x1>= -1')
plt.xlabel('x1')
plt.ylabel('x2')
plt.xlim(-4,4)
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Fesabile region bounded by constraints")
plt.legend()
plt.show()

obj = (x1_feas - 3)**2 + (x2_feas - 3)**2 
fig = plt.figure() 
plt.plot(x1_feas,obj)
plt.xlim(-4,4)
plt.xlabel("x1")
plt.ylabel("Objective function")
plt.title("Objective function value for feasible region")
plt.show()
```

```python
a1=np.linspace(-2,2,2000)
a2=a1
a3=a2

d1=np.linspace(-2,2,15)

x1 = np.linspace(-2, 2, 30)
x2 = np.linspace(-2, 2, 30)

X1, X2 = np.meshgrid(x1, x2)
X3 = 1 - X1 - X2

T,U,V=np.meshgrid(d1,d1,d1)

S4_x = T + U
S4_y = T + V
S4_z = T + (1 - U - V)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.plot(a1, a2, a3, color='red', label= 'S1: x1 = x2 = x3')
ax.plot_surface(X1, X2, X3, edgecolor='orange', label= 'S2: x1 + x2 + x3 = 1')
ax.plot(1/3,1/3,1/3,color='blue',label= 'S3 = S1S2=(1/3,1/3,1/3)')
ax.text(1/3,1/3,1/3,"+",color='blue')


ax.scatter(S4_x.flatten(),
           S4_y.flatten(),
           S4_z.flatten(),
           alpha=0.05,
           color='green',
           label='S4 = S1  S2')

ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("x3", labelpad=25)

ax.legend()
ax.view_init(elev=10, azim=120)
plt.tight_layout()
plt.show()


```

```python
x1 = np.linspace(-2, 2, 300)
x2 = np.linspace(-2, 2, 300)
X1, X2 = np.meshgrid(x1, x2)
obj=X1**2 + X2**2
fig=plt.figure()
ax=fig.add_subplot(111, projection='3d')

ax.plot_surface(X1, X2, obj,
                edgecolor='orange')

ax.view_init(elev=10, azim=120)
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("f(x1,x2)")
plt.show()
```

```python
x1 = np.linspace(0.1, 4, 300)
x2 = np.linspace(0.1, 4, 300)
X1, X2 = np.meshgrid(x1, x2)
obj= -X1*np.log(X1) -X2*np.log(X2)
fig=plt.figure()
ax=fig.add_subplot(111, projection='3d')

ax.plot_surface(X1, X2, obj, cmap='viridis',
                edgecolor='orange')

ax.view_init(elev=30, azim=45)
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("f(x1,x2)")
plt.show()
```
